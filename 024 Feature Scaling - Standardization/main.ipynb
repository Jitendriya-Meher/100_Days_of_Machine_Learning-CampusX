Standardization transforms numerical features to have a mean of 0 and a standard deviation of 1, aiding in comparing and interpreting features with different scales.

Standardization is a feature scaling technique that transforms numerical features so that they have:

Mean = 0

Standard Deviation = 1

(z-score)

It helps many machine learning models (like SVM, KNN, Logistic Regression) perform better by putting all features on a similar scale.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()

scalar
```

```python
data = scalar.fit_transform(df)
```

```python
df = pd.DataFrame(data, columns=df.columns)

df
```

```python
sns.scatterplot(
    df['EstimatedSalary']
)
```

```python
sns.scatterplot(
    y=df['EstimatedSalary'],
    x=df['Age'],
    hue=df['Purchased']
)
```
