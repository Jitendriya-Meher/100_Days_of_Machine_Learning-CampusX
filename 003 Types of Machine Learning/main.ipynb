1. Supervised Learning
What it is: The model is trained on labeled data (i.e., data that includes both inputs and the correct outputs).

Goal: Learn a mapping from inputs to outputs so it can predict outputs for new, unseen inputs.

Examples:

Predicting house prices based on size and location.

Email spam detection.

Diagnosing diseases from patient data.

✅ Common Algorithms:

Linear Regression

Logistic Regression

Decision Trees

Support Vector Machines (SVM)

Neural Networks

2. Unsupervised Learning
What it is: The model is given unlabeled data and must find hidden patterns or groupings on its own.

Goal: Discover the structure of the data.

Examples:

Customer segmentation for marketing.

Anomaly detection (e.g., fraud detection).

Topic modeling from text data.

✅ Common Algorithms:

K-Means Clustering

Hierarchical Clustering

Principal Component Analysis (PCA)

Autoencoders

3. Reinforcement Learning
What it is: The model (called an agent) learns by interacting with an environment, receiving feedback in the form of rewards or penalties.

Goal: Learn a strategy (policy) to maximize long-term rewards.

Examples:

Training a robot to walk.

Game-playing AIs (e.g., AlphaGo).

Autonomous driving.

✅ Core Concepts:

Agent, Environment, Actions, Rewards

Exploration vs. Exploitation

Q-Learning, Deep Q-Networks (DQN), Policy Gradients
